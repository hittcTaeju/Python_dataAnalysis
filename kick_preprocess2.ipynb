```python
import pandas as pd
user_parking = pd.read_csv('마포구_견인정보.csv')# 마포구에 21개의 지정 주차 구역이 있음
user_parking
```

```python
# '주소' 열에서 빈 문자열 또는 NaN 값이 있는지 확인
invalid_addresses = user_parking[user_parking['주소'].isna() | (user_parking['주소'] == '')]

if not invalid_addresses.empty:
    print("이상한 주소가 존재합니다:")
    print(invalid_addresses)
else:
    print("이상한 주소가 없습니다.")
```

```python
import kakao_geocoding

# '위도' 및 '경도' 열 추가 초기는 nan
user_parking['위도'] = float('nan')
user_parking['경도'] = float('nan')
count=0
for index, row in user_parking[10001:12000].iterrows():
  count+=1
  if count%1000==0:
    print(index)
  try:
    long_lat = kakao_geocoding.addr_to_lat_lon(str(row['주소']))
    user_parking.at[index, '위도'] = long_lat[1]
    user_parking.at[index, '경도'] = long_lat[0]
  except Exception as error:
    print(error)
user_parking
```
