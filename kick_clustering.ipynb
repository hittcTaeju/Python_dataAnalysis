```python
import pandas as pd
mapo_parking_lot = pd.read_csv('마포구_주차장.csv',encoding ='cp949')# 마포구에 21개의 지정 주차 구역이 있음
mapo_parking_lot
```

```python
user_parking = pd.read_csv('마포구_위경도추가.csv') #주차정보는 총 2382개
user_parking
```

```python
test_sample = user_parking[:30]
import matplotlib.pyplot as plt

import seaborn as sns
plt.figure(figsize=(10,8))
sns.scatterplot(data=test_sample,x='경도',y='위도')

plt.show()
```

```python
test_sample = test_sample[['위도', '경도']]

from sklearn.cluster import KMeans

Kmean = KMeans(n_clusters=5)
Kmean.fit(test_sample)
```

```python
centers = Kmean.cluster_centers_
centers
```

```python
plt.scatter(
    Kmean.cluster_centers_[:, 1], Kmean.cluster_centers_[:, 0],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
sns.scatterplot(data=test_sample,x='경도',y='위도')
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```
