```python
import pandas as pd
mapo_df= pd.read_csv('mapo_df.csv')
mapo_df
```

```python
correlation_matrix = mapo_df[['세대수', '남자인구수', '행정통수', '1인가구', '견인수']].corr()
correlation_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature 간의 상관 행렬')
plt.show()
```
