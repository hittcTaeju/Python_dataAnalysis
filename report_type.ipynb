```python
from pandas import *
parking_locations = read_csv("사용자_주차.csv")
parking_locations = parking_locations[parking_locations['주소'].str.contains('마포')]

keywords = ["버스정류장", "자전거", "점자블록", "지하철역", "차도", "통행장애", "횡단보도", "보도"]
dongs = ['공덕동', '구수동', '노고산동', '당인동', '대흥동', '도화동', '동교동', '마포동', '망원동', '상수동', '상암동', '서교동', '성산동', '신공덕동', '신수동', '신정동', '아현동', '연남동', '염리동', '용강동', '중동', '창전동', '토정동', '하중동', '합정동', '현석동']

parking_types = parking_locations['유형'] # 데이터셋 12184개(마포구 전체)
parking_dong = parking_locations['주소']

parking_types
```

```python
classified = {keyword : 0 for keyword in keywords} # 신고유형만 분류
dong = {dong_name : classified.copy() for dong_name in dongs} # 동에 따른 신고유형 분류

for types, address in zip(parking_types, parking_dong): 
    for keyword in keywords:
        if keyword in types:
            dong_name = address.split()[2]
            classified[keyword] += 1
            dong[dong_name][keyword] += 1
            break

classified
```

```python
total_sum = len(parking_types) # 총 신고수
for i, j in dong.items(): # 동별 신고유형 개수
    dong_sum = sum(j.values()) # 동별 신고 수 총합
    print(i + "\t ->  ", end="")
    print(*[f"{key} : {value} ({value/dong_sum*100:.2f}%)" for key, value in j.items()], sep="\t", end=f"\t총합 : {dong_sum} ({dong_sum/total_sum*100:.2f}%)")
    print()

classified_sum = sum(classified.values())

print("\n총 데이터 셋 :", total_sum)
print("분류된 데이터 셋 :", classified_sum, "\t분류되지 않은 데이터 셋(기타 항목) :", total_sum - sum(classified.values()))
```
