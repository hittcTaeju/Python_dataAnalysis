```python
import pandas as pd
mapo_parking_lot = pd.read_csv('마포구_주차장.csv',encoding ='cp949')# 마포구에 21개의 지정 주차 구역이 있음
mapo_parking_lot
```

```python
user_parking = pd.read_csv('마포구_위경도추가.csv') #주차정보는 총 2382개
user_parking
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
plt.figure(figsize=(10,8))
sns.scatterplot(data=user_parking,x='경도',y='위도')

plt.savefig('scatter_plot.png')
plt.show()
```

```python
import folium

mapo_map = folium.Map(location=[37.576494,	126.897538],zoom_start=15)
mapo_map
```

```python
for index, row in user_parking.iterrows():
  try:
    marker = folium.Marker([row['위도'],row['경도']])
    marker.add_to(mapo_map)
  except Exception as error:
    print(error)
mapo_map
```

```python
mapo_map.save('mapo_map.html')
```
