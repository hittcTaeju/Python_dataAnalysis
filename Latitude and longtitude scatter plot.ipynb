```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('마포구_위경도추가.csv')
```

```python
print(df.head())
```

```python
longitude = df['경도']  # '경도' 是韩文，意思是经度
latitude = df['위도']  # '위도' 是韩文，意思是纬度

# 创建散点图
plt.scatter(longitude, latitude, s=5)
plt.title('Latitude and longtitude scatter plot')
plt.xlabel('Longtitude')
plt.ylabel('Latitude')
plt.show()
```
