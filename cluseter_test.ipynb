```python
import pandas as pd
pd1 = pd.read_csv('real1~1000.csv')
pd2 = pd.read_csv('real1001~2000.csv')
pd3 = pd.read_csv('real2001~4000.csv')
pd4 = pd.read_csv('real4001~6000.csv')
pd5 = pd.read_csv('real6001~8000.csv')
pd6= pd.read_csv('real8001~10000.csv')
pd7=pd.read_csv('real10001~12000.csv')

total_pd = pd.concat([pd1, pd2, pd3,pd4,pd5,pd6,pd7], ignore_index=True)
```

```python
total_pd
```

```python
total_pd = total_pd[['위도', '경도']].dropna()
```

```python
total_pd
```

```python
import matplotlib.pyplot as plt

import seaborn as sns
plt.figure(figsize=(5,4))
sns.scatterplot(data=total_pd,x='경도',y='위도')

plt.show()
```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=50)
Kmean.fit(total_pd)
```

```python
centers = Kmean.cluster_centers_
centers
```

```python
# 각 데이터 포인트에 할당된 클러스터 레이블 얻기
labels = Kmean.labels_

# 각 클러스터 레이블의 개수 세기
cluster_counts = pd.Series(labels).value_counts()
```

```python
cluster_counts
```

```python

sns.scatterplot(data=total_pd,x='경도',y='위도')

plt.scatter(
    Kmean.cluster_centers_[:, 1], Kmean.cluster_centers_[:, 0],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```python
import folium

mapo_map = folium.Map(location=[37.546494,	126.937538],zoom_start=15)

mapo_map
```

```python
import folium

for row in centers:
    try:
        marker = folium.Marker([row[0], row[1]])  #  the first column is '위도' and the second column is '경도'
        marker.add_to(mapo_map)
    except Exception as error:
        print(error)
```
